```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.manifold import TSNE
import pickle
import pandas as pd
from autoencoders import metrics
```

# Analysis of the results of Experiment 1

 - **Experiment 1.** Ability for autoencoder clustering (AEC) to determine the source mechanism from a spectrogram.
    
    - Double couple, compensated linear vector dipole, and isotropic impulse sources will be used to generate synthetic seismograms with random nodal plane orientations
    
## See the end of this notebook for summary

```python
## Import the summary stats from the Sacred database

from pymongo import MongoClient
client = MongoClient()
db = client['seismicautoencoders']

info = {}
acc = []; nmi = []; ari = []; silhouette = []; algorithm = []; skips = []
runs = [16,17,18,19]
for run in runs:
    acc.append(db.metrics.find_one({'run_id': run, 'name':'accuracy_3'})['values'][-1])
    nmi.append(db.metrics.find_one({'run_id': run, 'name':'NMI_3'})['values'][-1])
    ari.append(db.metrics.find_one({'run_id': run, 'name':'ARI_3'})['values'][-1])
    silhouette.append(db.metrics.find_one({'run_id': run, 'name':'silhouette score'})['values'][-1])
    config = db.runs.find_one({'_id': run})['config']
    algorithm.append(config['algorithm'])
    skips.append(config['skips'])
info['algorithm'] = algorithm
info['skips'] = skips
info['acc'] = acc
info['nmi'] = nmi
info['ari'] = ari
info['silhouette'] = silhouette


run_info = pd.DataFrame(info)
run_info.to_csv('Experiment1/clustering_metrics_results.csv')
run_info
```

## Look at each individual algorithm

```python
algos = ['causal_skipsTrue','causal_skipsFalse','acausal_skipsTrue','acausal_skipsFalse']

algo = algos[3]

save_dir = 'Experiment1/{}'.format(algo)

#runs = [16]

with open('Experiment1\synthetics.pkl','rb') as f:
    synthetics = pickle.load(f)
    
lvs_causal_skips_3 = np.load('Experiment1/{}/3clusters/latent_vectors_3_clusters.npy'.format(algo))
```

```python
mech2cat = {'DC':0, 'CLVD':1, 'Isotropic':2} #assigns numeric value to mechanism type

#have to convert to numpy arrays lists to put into a dataframe
seismograms = synthetics['seismograms'].tolist()
spectrograms = synthetics['spectrograms'].tolist()
moment_tensors = synthetics['moment_tensors']

results = synthetics
results['cluster_assignments'] = np.load('Experiment1/{}/3clusters/cluster_assignments_3_clusters.npy'.format(algo))
results['source_mech_cat'] = [mech2cat[m] for m in results['source_mechanisms']]

del results['seismograms']
del results['spectrograms']
del results['moment_tensors']
```

```python
results = pd.DataFrame(results)
results.to_csv('{}/results.csv'.format(save_dir))
```

## Let's look at the clustering accuracy for only DC and CLVD sources, since isotropic sources cluster perfectly

```python
alg = []; acc2 = []; nmi2 =[]; ari2 = []

for a in algos:

    df = pd.read_csv('Experiment1/{}/results.csv'.format(a))
    df2 = df[(df.source_mech_cat==0) | (df.source_mech_cat==1)]
    
    y_true = df2.source_mech_cat
    y_pred = df2.cluster_assignments
    
    acc2.append(metrics.acc(y_true, y_pred))
    nmi2.append(metrics.nmi(y_true, y_pred))
    ari2.append(metrics.ari(y_true, y_pred))
    
results2 = pd.DataFrame({'algorithm':algos,'acc':acc2, 'nmi':nmi2, 'ari':ari2})  
results2
```

```python
run_info.to_csv('Experiment1/clustering_metrics_results.csv')
results2.to_csv('Experiment1/only_DCandCLVD_clustering_metrics_results.csv')
```

 - Skip connections are slightly worse, by a fraction of a percent.  Unknown if significant.

 - Acasual with skip connections fares slightly better on the mutual information and adjusted rand index scores.
