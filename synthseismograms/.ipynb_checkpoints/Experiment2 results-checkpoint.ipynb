# Analysis of the results of Experiment 2

 - **Experiment 2.** Ability for AEC to determine the *source time function* class 
      - Lognormal, boxcar, and triangular source time functions will be used to generate seismograms with a single source mechanism type (double couple), random nodal plane orientation, and varying durations (1 - 10 seconds).
      
      
Will want to analyze the distribution of accuracies for each algorithm, pick a run fdor each algorithm and see how they were clustered in terms of stf, nodal plane, duration.

## Import the summary stats from the Sacred database

```python
from pymongo import MongoClient
client = MongoClient()
db = client['seismicautoencoders']

info = {}
run_ids = []; acc = []; nmi = []; ari = []; silhouette = []; algorithm = []; skips = []; L=[]; Lr=[]; Lc=[]

causal_skip_runs = [20,24,28,29,30,31,32,33,34,35,36]
causal_noskip_runs = [21,25,37,38,39,40,41,42,43,44,45]
acausal_skip_runs = [22,26,46,47,48,49,50,51,52,53,54]
acausal_noskip_runs = [23,27,55,56,57,58,59,60,61,62,63]

for c in db.runs.find({'experiment.name':'Experiment2', 'status':'COMPLETED'}):#, 'config.latent.dims':5}):
    run = c['_id']
    run_ids.append(run)
    acc.append(db.metrics.find_one({'run_id': run, 'name':'accuracy_3'})['values'][-1])
    nmi.append(db.metrics.find_one({'run_id': run, 'name':'NMI_3'})['values'][-1])
    ari.append(db.metrics.find_one({'run_id': run, 'name':'ARI_3'})['values'][-1])
    L.append(db.metrics.find_one({'run_id': run, 'name':'total loss_3'})['values'][-1])
    Lr.append(db.metrics.find_one({'run_id': run, 'name':'Lr(reconstruction loss_3'})['values'][-1])
    Lc.append(db.metrics.find_one({'run_id': run, 'name':'Lc(clustering loss_3'})['values'][-1])
    silhouette.append(db.metrics.find_one({'run_id': run, 'name':'silhouette score'})['values'][-1])
    config = db.runs.find_one({'_id': run})['config']
    algorithm.append(config['algorithm'])
    skips.append(config['skips'])  

info['run'] = run_ids
info['algorithm'] = algorithm
info['skips'] = skips
info['acc'] = acc
info['nmi'] = nmi
info['ari'] = ari
info['L'] = L
info['Lc'] = Lc
info['Lr'] = Lr
info['silhouette'] = silhouette

run_info = pd.DataFrame(info)
run_info.to_csv('Experiment2/clustering_metrics_results.csv')
#run_info
```

```python
#find the run # with best accuracy
# maybe I should do this by saving runs with minimum loss?

def saveBestRun(algorithm, skips, save_dir):
    #i = run_info[(run_info['algorithm']==algorithm) & (run_info['skips']==skips)].acc.idxmax()
    i = run_info[(run_info['algorithm']==algorithm) & (run_info['skips']==skips)].L.idxmin()
    print(i)
    max_run = run_info.loc[i].run
    run = db.runs.find_one({'_id':int(max_run)})
    #print(run['artifacts'])
    for f in run['artifacts']:
        #print(f['name'])
        chunk = db.fs.chunks.find_one({'files_id':f['file_id']})['data']
        #print(len(chunk))
        with open(save_dir+'/'+f['name'], 'wb') as f:
            f.write(chunk)
```

### Get the model weights from MongoObserver

```python
def listMongoObserverArtifacts(run_id, show=True):
    '''
    Get a list of names of MongoObserver artifacts from a particular run.
    
    :param run_id: the id number of the run 
    :param show: default True, whether to print the names and index values of the artifacts
    
    :return A list of the artifact names, ordered the same as in the MongoDB database'''
    names = []
    for i,artifact in enumerate(db.runs.find_one({'_id':110})['artifacts']):
        art_name = artifact['name']
        names.append(art_name)
        if show: print('{}: {}'.format(i,art_name))
    return names
    
def getMongoObserverArtifact(run_id, file_num):
    """
    Get an artifact file from a Sacred run stored in MongoDB
    
    :param run_id: The id number of the run.
    :param file_num: Index of the file in the artifact list stored in MongoDB.  Can be found with listMongoObserverArtifacts.
    
    :return the data in the artifact.
    """
    run = db.runs.find_one({'_id':run_id})
    file_id = run['artifacts'][file_num]['file_id']
    chunk = db.fs.chunks.find_one({'files_id':file_id})['data']
    return chunk
                            
```
